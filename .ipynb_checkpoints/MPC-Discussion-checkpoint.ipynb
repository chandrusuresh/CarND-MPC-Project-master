## Experiments with delayed control input without latency in system

The reference speed was set to 30 meters per second. At slow speed, the car would definitely be able to drive and tuning the parameters would not lead to any significant changes to the vehicle response/tracking. 30m/s is appropriate to navigate the lake track with some sharp corners.

For this speed, it was found that applying the current predicted throttle and steer commands would lead to a zig-zag motion as the car is trying to correct its position & orientation at every step. Note that this zig-zag motion is also a function of the weights set to each of the cost functions.

The gains of the cost function therefore is first tuned to keep the car on track tracking 30m/s.

An MPC is designed to predict the control actions over a horizon of 50 time steps. 50 time steps at 0.01 sec time step equates to 0.5 sec into the future, which is a long enough to account for most latencies the system and sensors might cause.

## Cost Function
The cost function is defined by 7 terms as described below. Each term is given a weight wi. The choice of wi is discussed here for the case to keep the car on track at 30m/s. The rest of the experiments keep these weights as a baseline and modified (and documented) as appropriate.
Initialization was a little bit of a challenge at 30m/s, since the car tries to get to full throttle even when its orientation is a little off the reference trajectory. Since the initial epsi is large, the car tries to correct for it and leads to oscillatory behavior. These oscillations at a speed of 30m/s lead to instability and leads to the car leaving the track. To fix this, the weights w4 was set to 100 and w6 was set to 10. This means that the controller tries to keep the steering wheel input (w4) and its rate (w6) as low as possible. Thus, w4=100, w6=10 was sufficient to complete a lap at 30m/s.

The cost function with weight w1 serves to mimimize the cross track error, while the one with weight w2 minimizes the error in heading angle from the reference trajectory. The term with weight w3 serves to follow the reference speed while w6 and w7 place emphasis on the rate of change of the control inputs, to prevent large changes in control input from one time step to another. The terms with weights w4 & w5 serve to minimize the magnitude of the control inputs as much as possible.

The MPC then computes the control action that minimizes the cost function over the prediction horizon. The control horizon is kept at 1 time step. To incorporate latencies, the control actions are held constant at the values from the previous time step during the period of latency.

The following report details the tuning of the weights and how the latency is handled.

## Experiments to reduce CTE
The above weights on w4 and w6 were good to keep the car on the track, but the CTE was upto 1.5 meters. This is not good for urban driving. A few experiments were performed to reduce the peak CTE to within 0.2 meters. This can be done by changing w1. While increasing w1 will improve the worst CTE, a really high w1 will lead to aggresive corrections on the steering wheel angle and oscillatory motion. Therefore, w1 value was set to 2.
Another way to reduce the max CTE is by actuating the steering and throttle solution from the mpc a few time steps into the future.

The best solution to reduce CTE was to apply the throttle and steering 2 time steps into the future. The side effect of this is also a smooth steering trace with a marginal improvement to the CTE.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

cwd = os.getcwd()
outPath = os.path.join(cwd,"build")
files = os.listdir(outPath)

csv_files = ['MPC_refVel_30.000000_inputDelayed_0.csv', 'MPC_refVel_30.000000_inputDelayed_0_w1_2.csv',
             'MPC_refVel_30.000000_inputDelayed_2.csv']

fig, ax = plt.subplots(3,2,figsize=(20,10))
lgd = []

for i in range(len(csv_files)):
    df = pd.read_csv(os.path.join(outPath,csv_files[i]))
    nPts = np.divide(range(len(df["x"])),float(len(df["x"])))
    ax[0,0].plot(df["x"],df["y"])
    ax[0,1].plot(nPts,df["v"])
    ax[1,0].plot(nPts,df["epsi"])
    ax[1,1].plot(nPts,df["cte"])
    ax[2,0].plot(nPts,df["throttle"])
    ax[2,1].plot(nPts,df["steer"]) 
    
    ax[0,0].set_title("Vehicle Path")
    ax[0,1].set_title("Speed")
    ax[1,0].set_title("Error in heading")
    ax[1,1].set_title("CTE")
    ax[2,0].set_title("Throttle")
    ax[2,1].set_title("Steer")

ax[2,0].legend(['Baseline (BL)', 'BL & w1=2', 'BL & input feedfwd by 2 future time steps'])
plt.show()
```

## 100ms Latency
The baseline weights were simulated with a 100ms latency to simulate actuator dynamics. Based on feedback from the first review, I changed my code such that the control inputs are constrained (in the optimization) to be the same as that of the previous time step during the period of the latency.

I reduced the speed to 20mph and modified the following weights: w1 = 2, w2 = 5 and w3 = 0.1

w1 was increased to 2 because the latency introduced some CTE at the turns. w2 was set to 5 to reduce zig zag motions and w3 was set to 0.1 to allow the controller some tolerance with speed so that there was not a lot of chattering in the throttle trace.

Application of the predicted input 10 time steps (0.1 sec), the time of latency into the future is sufficient for the car to complete a lap around the track.

```python
del csv_files[1]
csv_files.append('MPC_refVel_20.000000_inputDelayed_0_latency_thrTuning.csv')

fig, ax = plt.subplots(3,2,figsize=(20,10))
lgd = []

for i in range(len(csv_files)):
    df = pd.read_csv(os.path.join(outPath,csv_files[i]))
    nPts = np.divide(range(len(df["x"])),float(len(df["x"])))
    ax[0,0].plot(df["x"],df["y"])
    ax[0,1].plot(nPts,df["v"])
    ax[1,0].plot(nPts,df["epsi"])
    ax[1,1].plot(nPts,df["cte"])
    ax[2,0].plot(nPts,df["throttle"])
    ax[2,1].plot(nPts,df["steer"]) 
    
    ax[0,0].set_title("Vehicle Path")
    ax[0,1].set_title("Speed")
    ax[1,0].set_title("Error in heading")
    ax[1,1].set_title("CTE")
    ax[2,0].set_title("Throttle")
    ax[2,1].set_title("Steer")

ax[0,1].legend(['Baseline', 'BL & input delay by 2 steps', 'BL & input feed fwd by 10 steps & 100ms latency'])
plt.show()
```

While the car is able to complete a lap of the lake loop with a 100ms latency, the throttle trace is much more smooth than the previous submission.
